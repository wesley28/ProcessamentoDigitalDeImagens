```python
import random

class Imagem:

    def __init__(self, largura, altura):
        self.largura = largura
        self.altura = altura
        self.pixels = [[[0, 0, 0] for i in range(altura)] for j in range(largura)]

    def get_pixel(self, x, y):
        return self.pixels[x][y]

    def set_pixel(self, x, y, valor):
        self.pixels[x][y] = valor

    def salvar(self, arquivo):
        with open(arquivo, 'w') as f:
            f.write('P3\n')
            f.write(str(self.largura) + ' ' + str(self.altura) + '\n')
            f.write('255\n')
            for linha in self.pixels:
                for pixel in linha:
                    f.write(' '.join(str(int(pixel * 16)) for pixel in linha) + '\n')

    def mostrar(self):
        import matplotlib.pyplot as plt

        plt.imshow(self.pixels, cmap='gray')
        plt.show()

imagem = Imagem(100, 100)

for x in range(imagem.largura):
    for y in range(imagem.altura):
        for i in range(3):
            valor = random.randint(0, 15)
            imagem.set_pixel(x, y, valor)

imagem.salvar('imagem.pbm')

imagem.mostrar()
```
