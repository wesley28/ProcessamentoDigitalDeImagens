```python
class Imagem:

    def __init__(self, largura, altura, profundidade):
        self.largura = largura
        self.altura = altura
        self.profundidade = profundidade
        self.pixels = [[0 for i in range(altura)] for j in range(largura)]

    def get_pixel(self, x, y):
        return self.pixels[x][y]

    def set_pixel(self, x, y, valor):
        self.pixels[x][y] = valor

    def salvar(self, arquivo):
        with open(arquivo, 'w') as f:
            f.write('P2\n')
            f.write(str(self.largura) + ' ' + str(self.altura) + '\n')
            f.write(str(self.profundidade) + '\n')
            for linha in self.pixels:
                f.write(' '.join(str(x) for x in linha))
                f.write('\n')

    def mostrar(self):
        import matplotlib.pyplot as plt

        plt.imshow(self.pixels, cmap='gray')
        plt.show()

# Crie uma imagem de 100x100 pixels com profundidade de 16 bits
imagem = Imagem(100, 100, 16)

# Defina os pixels da imagem para valores aleat√≥rios
for x in range(imagem.largura):
    for y in range(imagem.altura):
        imagem.set_pixel(x, y, random.randint(0, 15))

# Salve a imagem em um arquivo chamado `imagem.pgm`
imagem.salvar('imagem.pgm')

# Mostre a imagem na tela
imagem.mostrar()
```
